# Data Visualisation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./bank/bank_etl.csv")
df.head()
```

### Education type, inline with term deposit


```python
fig = plt.figure(figsize=(8, 3))
sns.countplot(data=df, y="education", hue='y')
plt.xlabel("Education Type")
plt.ylabel("Education Counts")
plt.title("Counts of education, inline with term deposit", pad=15,)
handles, labels = plt.gca().get_legend_handles_labels() # get current labels and handles
new_labels = ['No Term Deposit', 'Term Deposit']
plt.legend(handles, new_labels)
plt.show()
```

### show age, with marital status, inline with hue of y

```python
fig = plt.figure(figsize=(8, 3))
sns.barplot(data=df, x="age", y='marital', hue='y')
plt.xlabel("Age")
plt.ylabel("Marital Status")
plt.title("Age, with marital status, inline with hue of term deposit", pad=15,)
handles, labels = plt.gca().get_legend_handles_labels() # get current labels and handles
new_labels = ['No Term Deposit', 'Term Deposit']
plt.legend(handles, new_labels)
plt.show()
```

### show age, with job status, inline with hue of y

```python
fig = plt.figure(figsize=(8, 6))
sns.barplot(data=df, x="age", y='job', hue='y' )
plt.xlabel("Age")
plt.ylabel("Job Status")
plt.title("Age, with job status, with hue of y", pad=15,)
handles, labels = plt.gca().get_legend_handles_labels() # get current labels and handles
new_labels = ['No Term Deposit', 'Term Deposit']
plt.legend(handles, new_labels)
plt.show()
```

### number of marital status, if the signed up for a term deposit

```python
fig = plt.figure(figsize=(8, 4))
sns.countplot(data = df, x='marital', hue='y' )
plt.xlabel("Counts")
plt.ylabel("Marital Status")
plt.title("Number of marital statuses inline with a term deposit", pad=15,)
handles, labels = plt.gca().get_legend_handles_labels() # get current labels and handles
new_labels = ['No Term Deposit', 'Term Deposit']
plt.legend(handles, new_labels)
plt.show()
```

### number of customers in jobs type, that subscribed to a term deposit

```python
fig = plt.figure(figsize=(8, 5))
sns.countplot(data = df, y='job', hue='y' )
plt.xlabel("Counts")
plt.ylabel("Job")
plt.title("Number of customers in jobs type, that subscribed to a term deposit", pad=15)
handles, labels = plt.gca().get_legend_handles_labels() # get current labels and handles
new_labels = ['No Term Deposit', 'Term Deposit']
plt.legend(handles, new_labels)
plt.show()
```
